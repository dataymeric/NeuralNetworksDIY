```python
%cd ..
import numpy as np
import matplotlib.pyplot as plt
from src.activation import *

__all__ = [
    TanH(),
    Sigmoid(),
    StableSigmoid(),
    Softmax(),
    LogSoftmax(),
    ReLU(),
    LeakyReLU(),
    Softplus(),
]
```

```python
X = np.linspace(-5, 5, 500)

def plot_activation(activation, data):
    plt.plot(data, activation(data))
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(f"{activation.__class__.__name__}(x)")
    plt.show()

```

```python
for activation in __all__:
    plot_activation(activation, X)
```

```python
plt.plot(X, LeakyReLU()(X), label=r"$\alpha=0.01$")
plt.plot(X, LeakyReLU(alpha=0.1)(X), label=r"$\alpha=0.1$")
plt.xlabel("x")
plt.ylabel("y")
plt.title("LeakyReLU(x)")
plt.legend()
plt.show()
```
